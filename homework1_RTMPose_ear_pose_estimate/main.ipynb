# 图片关键点检测

将自己耳朵的图片和视频放入 ./mmpose/data 文件夹中

将 detection 文件夹中转换好的 checkpoint 复制到 mmpose 对应文件夹

由于我的测试视频与训练集相差较大，所以检测这块阈值调低一点，要不然检测不到了。

```python
# RTMDet --> RTMPose
!python demo/topdown_demo_with_mmdet.py \
        data/rtmdet_tiny_ear.py \
        checkpoint/rtmdet_tiny_ear_epoch_120.pth \
        data/rtmpose-s-ear.py \
        checkpoint/rtmpose-s-ear_epoch_300.pth \
        --input data/myear.jpg \
        --output-root outputs/G2_RTMDet-RTMPose \
        --device cuda:0 \
        --bbox-thr 0.2 \
        --kpt-thr 0.2 \
        --nms-thr 0.3 \
        --radius 10 \
        --thickness 10 \
        --draw-bbox \
        --draw-heatmap \
        --show-kpt-idx
```

# 视频关键点检测

```python
# RTMDet --> RTMPose
!python demo/topdown_demo_with_mmdet.py \
        data/rtmdet_tiny_ear.py \
        checkpoint/rtmdet_tiny_ear_epoch_120.pth \
        data/rtmpose-s-ear.py \
        checkpoint/rtmpose-s-ear_epoch_300.pth \
        --input data/myear.mp4 \
        --output-root outputs/G2_Video \
        --device cuda:0 \
        --bbox-thr 0.25 \
        --kpt-thr 0.2 \
        --nms-thr 0.3 \
        --radius 10 \
        --thickness 10 \
        --draw-bbox \
        --draw-heatmap \
        --show-kpt-idx
```
